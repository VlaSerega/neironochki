# Импорты

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Грузим данные

```python
cancer = load_breast_cancer()

X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.Series(cancer.target)
```

```python
X.describe()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tree_clf = DecisionTreeClassifier(random_state=10, max_depth=5)
tree_clf.fit(X_train, y_train)

pred = tree_clf.predict(X_test)
print('Test acc:', accuracy_score(y_test, pred))
```

```python
plt.figure(figsize=(40, 20))
plot_tree(tree_clf, 
          filled=True, 
          feature_names=cancer.feature_names, 
          class_names=cancer.target_names.tolist())
plt.title("Decision Tree")
plt.show()
```

## Descisionboundary

```python
def decisionboundary(model, X, y):
    x_min = X[:, 0].min()
    x_max = X[:, 0].max()
    y_min = X[:, 1].min()
    y_max = X[:, 1].max()
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.6)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.jet)


X_train_2 = X_train[["worst area", "mean texture"]]
tree_clf.fit(X_train_2, y_train)

plt.figure(figsize=(40, 20))
decisionboundary(tree_clf, X_train_2.values, y_train)
```
